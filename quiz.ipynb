```python
def quiz():
    questions = [
        {"q": "Capital of India?", "a": "delhi"},
        {"q": "5 + 7 = ?", "a": "12"},
        {"q": "Python is a (language/animal)?", "a": "language"},
        {"q": "Sun rises from (east/west)?", "a": "east"}
    ]

    score = 0

    print("🎮 Welcome to Quiz Game!")

    for i, item in enumerate(questions, start=1):
        ans = input(f"Q{i}: {item['q']} ").lower()

        if ans == item["a"]:
            print("✅ Correct!")
            score += 1
        else:
            print("❌ Wrong!")

    print(f"\n🏁 Final Score: {score}/{len(questions)}")

quiz()
```
